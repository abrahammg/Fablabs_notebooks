## Practice: Creating a Simple RNN for Temperature Prediction

In this example, we will use a simple Recurrent Neural Network (RNN) to predict daily temperatures based on past records. RNNs are useful for time-series forecasting, such as weather prediction, stock market analysis, and trend detection.


## 1. Importing Required Libraries

We import NumPy for numerical operations, TensorFlow/Keras to build and train the RNN, Matplotlib for visualization, and Scikit-learn for scaling.


Nesta celda importamos as librarías necesarias para traballar con RNNs:

- **numpy**: Para manipulación numérica con arrays.
- **tensorflow** e **keras**: Para definir e adestrar a RNN.
- **matplotlib.pyplot**: Para visualizar datos e resultados.
- **sklearn.preprocessing.MinMaxScaler**: Para escalado de datos.

A mensaxe confirmará que as importacións foron correctas.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Input, SimpleRNN, Dense
from sklearn.preprocessing import MinMaxScaler
print("Libraries imported successfully!")
```

## 2. Load and Visualize Temperature Data

We load the daily temperature data from CSV and plot it to observe seasonal patterns.


Nesta celda cargamos e exploramos o conxunto de datos de temperatura diaria:

1. Usamos **pandas** para ler o ficheiro CSV `daily_temperature.csv`.
2. Extraemos as columnas `day` e `temperature`.
3. Plotamos a evolución da temperatura ao longo dos días para comprender a serie temporal.

```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/daily_temperature.csv')
days = df['day'].values
temperature = df['temperature'].values

# Plot
plt.figure(figsize=(10,4))
plt.plot(days, temperature, label='Temperature')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.title('Synthetic Daily Temperature (1 year)')
plt.legend()
plt.show()
```

## 3. Preparing the Data for the RNN

We normalize the temperatures, create sequences of the past 7 days as input, and the next day's temperature as output. Then split into train/test sets (80/20).


Nesta celda preprocesamos os datos:

1. **Escalado**: Transformamos as temperaturas ao rango [0,1] cun `MinMaxScaler`.
2. **Creación de secuencias**: Definimos `sequence_length=7` días e construímos pares (X, y) onde X son 7 pasos anteriores e y o seguinte.
3. **Split**: Separamos o 80% para adestramento e o 20% restante para proba.

Imprimimos o número de mostras de adestramento e test.

```python
# Normalize the data
scaler = MinMaxScaler(feature_range=(0,1))
temp_scaled = scaler.fit_transform(temperature.reshape(-1,1))

# Create sequences
sequence_length = 7
X, y = [], []
for i in range(len(temp_scaled) - sequence_length):
    X.append(temp_scaled[i:i+sequence_length])
    y.append(temp_scaled[i+sequence_length])
X, y = np.array(X), np.array(y)

# Split into train/test
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"Training samples: {X_train.shape[0]}, Testing samples: {X_test.shape[0]}")
```

## 4. Building and Training the RNN Model

We build a Sequential model with one SimpleRNN layer and one Dense output. We compile with Adam optimizer and MSE loss, then train for 50 epochs.


Nesta celda definimos, compilamos e adestramos a RNN:

- Comezamos cun modelo **Sequential** de Keras.  
- Empregamos unha capa **Input** explícita para indicar a forma da secuencia (sequence_length pasos, 1 característica).  
- Engadimos unha capa **SimpleRNN** con 10 unidades e activación **ReLU**.  
- Rematamos cunha capa **Dense(1)** para predicir o valor continuo seguinte.  
- Compilamos con **Adam** e perda **MSE**, e adestramos durante 50 épocas coa validación en datos de test.

```python
model = Sequential([
    Input(shape=(sequence_length, 1)),
    SimpleRNN(10, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), verbose=1)
```

## 5. Making Predictions

We predict on the test set, inverse-transform the results, and compare predicted vs actual temperatures.


Nesta celda avaliamos as predicións do modelo:

1. Usamos `model.predict` en `X_test` e invertimos o escalado.
2. Rescalamos tanto predicións como valores reais para obter graos en Celsius.
3. Plotamos a serie real fronte á predición con liñas sólidas e discontinuas.
4. Axustamos etiquetas, título e lenda para claridade.

```python
# Predict
predictions = model.predict(X_test)
pred_rescaled = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test)

# Plot results
plt.figure(figsize=(10,4))
plt.plot(y_test_rescaled, label='Actual Temperature')
plt.plot(pred_rescaled, label='Predicted Temperature', linestyle='dashed')
plt.xlabel('Time Step')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Prediction Using RNN')
plt.legend()
plt.show()
```
